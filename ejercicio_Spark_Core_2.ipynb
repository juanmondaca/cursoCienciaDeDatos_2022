```python
from pyspark import SparkContext
```

```python
sc = SparkContext.getOrCreate()
```

```python
# cargar un archivo de texto con sc
archivoLicenciaRDD = sc.textFile("C:\soft\spark-3.0.3-bin-hadoop2.7\LICENSE")
```

```python
archivoLicenciaRDD.count()
```

```python
archivoLicenciaRDD.collect()
```

```python
# realizar una transformacion (map)
archivoLicenciaSplitRDD = archivoLicenciaRDD.map(lambda linea : linea.split(" "))
archivoLicenciaSplitRDD.collect()
```

```python
# buscar la palabra Apache
lineasApacheRDD = archivoLicenciaRDD.filter(lambda l : "Apache" in l)
```

```python
lineasApacheRDD.collect()
```
